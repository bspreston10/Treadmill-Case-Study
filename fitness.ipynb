```python
sns.boxplot(data=df, x='Product', y='Fitness', hue='Product')
plt.title('Product vs Fitness')
```

The more fit the customer, the more likely he is to buy the KP781

```python
def prob_fitness(fitness):
    pd_fitness = pd.crosstab(index=df['Fitness'], columns=[df['Product']])
    prob_281 = pd_fitness['KP281'][fitness]/pd_fitness.loc[fitness].sum()
    prob_481 = pd_fitness['KP481'][fitness]/pd_fitness.loc[fitness].sum()
    prob_781 = pd_fitness['KP781'][fitness]/pd_fitness.loc[fitness].sum()
    print(f"P(KP281/{fitness}): {prob_281:.2f}") 
    print(f"P(KP481/{fitness}): {prob_481:.2f}")
    print(f"P(KP781/{fitness}): {prob_781:.2f}\n")
    
prob_fitness(1)
prob_fitness(2)
prob_fitness(3)
prob_fitness(4)
prob_fitness(5)
```

The probability of a customer buying the most expensive treadmill increases as fitness level increases. Between fitness level 1-3 they are basically indifferent when choosing the KPI281 and the KPI481.
