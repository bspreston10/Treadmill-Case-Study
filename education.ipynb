```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('aerofit_treadmill_data.csv')
```

```python
sns.displot(df['Education'], kde=True, color='green')
```

```python
sns.boxplot(data=df, x='Education', color='green', flierprops=dict(markerfacecolor='r'))
```

```python
sns.boxplot(data=df, x='Product', y='Education', hue='Product')
plt.title('Product vs Education')
```

People with higher educations tend to buy the KP781 more
