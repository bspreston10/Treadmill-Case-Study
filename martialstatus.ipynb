```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('aerofit_treadmill_data.csv')
```

```python
df['MaritalStatus'].value_counts()
```

```python
sns.countplot(data=df, x='Product', hue='MaritalStatus')
plt.title('Product and Marital Status')
```

```python
def prob_martialstatus(status):
    pd_ms = pd.crosstab(index=df['MaritalStatus'], columns=[df['Product']])
    prob_281 = pd_ms['KP281'][status]/pd_ms.loc[status].sum()
    prob_481 = pd_ms['KP481'][status]/pd_ms.loc[status].sum()
    prob_781 = pd_ms['KP781'][status]/pd_ms.loc[status].sum()
    print(f"P(KP281/{status}): {prob_281:.2f}") 
    print(f"P(KP481/{status}): {prob_481:.2f}")
    print(f"P(KP781/{status}): {prob_781:.2f}\n")
    
prob_martialstatus('Single')
prob_martialstatus('Partnered')
```
